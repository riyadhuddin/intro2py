```python
import matplotlib
from matplotlib import pyplot as plt, patches
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot(111)
rect = patches.Rectangle((2, 2), 10, 6, color='green')
circle = patches.Circle((6, 5), radius=2, color='red')
ax.add_patch(rect)
ax.add_patch(circle)
plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.axis('equal')
plt.show()
```
